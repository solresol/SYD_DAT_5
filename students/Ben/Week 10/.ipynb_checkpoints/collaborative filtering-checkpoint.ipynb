```python
import pandas
```

```python
import networkx
```

```python
brand = pandas.read_csv('../../../data/user_brand.csv')
```

```python
users = set()
stores = set ()
graph = networkx.Graph()
for (user, store) in zip(brand.ID, brand.Store):
    if user not in users:
        users.add(user)
        graph.add_node(user)
    if store not in stores:
        stores.add(store)
        graph.add_node(store)
    graph.add_edge(user,store)
```

```python
store_pairs = [(x,y) for x in stores for y in stores if x !=y]
```

```python
sorted(list(networkx.jaccard_coefficient(graph, store_pairs)), key = (lambda x:x[2]), reverse=True)
```

```python
def amazon_weight(store):
    networkx.degree(graph, [store])
amazon_weight('YSL')
```
