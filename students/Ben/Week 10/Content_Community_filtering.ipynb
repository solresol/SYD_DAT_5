```python
import pandas
import numpy
```

```python
data=pandas.read_csv('../../../data/songs.csv', index_col='name')
```

```python
listeners=pandas.read_csv('../../../data/listeners.csv', index_col='name')
```

```python
import sklearn.linear_model
```

```python
juanita_model=sklearn.linear_model.LinearRegression()
```

```python
juanita_model.fit(data[listeners.juanita.notnull()], listeners.juanita[listeners.juanita.notnull()])
```

```python
sorted(zip(juanita_model.predict(data),data.index), reverse=True)

# make the recommendation for the top results) 
```

```python
hugh_model=sklearn.linear_model.LinearRegression()
hugh_model.fit(data[listeners.hugh.notnull()], listeners.hugh[listeners.hugh.notnull()])
```

```python
sorted(zip(hugh_model.predict(data),data.index),reverse=True)
```
