```python
def is_a_real_word(x):
    return x not in ".,[]:"
```

```python
cleaned_up_hamlet = [
    x.lower()
        for x in nltk.corpus.gutenberg.words("shakespeare-hamlet.txt")
        if is_a_real_word(x)
]
cleaned_up_hamlet
```
