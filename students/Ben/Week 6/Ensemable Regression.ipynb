```python
import sklearn.datasets
```

```python
b = sklearn.datasets.load_boston()
```

```python
b.keys()
```

```python
b['feature_names']
```

```python
import pandas
```

```python
Boston = pandas.DataFrame(data=b['data'], columns=b['feature_names'])
```

```python
Boston['price']=b['target']
```

```python
Boston
```

```python
import sklearn.ensemble
```

```python
bag = sklearn.ensemble.BaggingRegressor(max_features=3, n_estimators=25, max_samples=100, n_jobs=1)
```

```python
boston_predictors = Boston.drop('price', axis=1)
boston_predictors
Boston[Boston.columns[:-1]]
```

```python
print (b['DESCR'])
```

```python
bag.fit(boston_predictors,Boston.price)
```

```python
bag.estimators_
```

```python
sklearn.tree.export_graphviz(bag.estimator_[0], .....)
```

```python
zip(bag.predict(boston_predictors), Boston.price) # compare predict with real price 
```

```python
%matplotlib inline 
```

```python
import matplotlib.pyplot
matplotlib.pyplot.scatter(Boston.price, bag.predict(boston_predictors))
```
