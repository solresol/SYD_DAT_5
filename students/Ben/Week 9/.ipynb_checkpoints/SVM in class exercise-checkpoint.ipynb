```python
import pandas
import numpy
%matplotlib inline
import matplotlib.pyplot
```

```python
import sklearn.datasets
```

```python
circles_xy, which_ring = sklearn.datasets.make_circles()
```

```python
matplotlib.pyplot.scatter(circles_xy[:,0], circles_xy[:,1],
                         c=which_ring, cmap=matplotlib.pyplot.cm.Blues)
```

```python
import sklearn.svm
```

```python
svc= sklearn.svm.SVC()
```

```python
svc.fit(circles_xy, which_ring)
```

```python
which_ring
```

```python
svc.predict([circles_xy[0]])
```

```python
svc.predict([circles_xy[1]])
```

```python
svc.predict([circles_xy[2]])
```

```python
sklearn.metrics.accuracy_score(svc.predict(circles_xy), which_ring)
```

```python
numpy.arange(-1.5, 1.5, 0.05) # find the points in steps of 0.05 between -1.5 to 1.5 
```

```python
grid = [(x,y) for x in numpy.arange(-1.5, 1.5, 0.05) for y in numpy.arange(-1.5, 1.5, 0.05)]
```

```python
len(grid)
```

```python
df = pandas.DataFrame(
    data = {
        'x':[t[0] for t in grid], 
        'y':[t[1] for t in grid]
    }
    )
```

```python
predictions = svc.predict(df)
```

```python
matplotlib.pyplot.scatter(df.x, df.y, c=predictions, cmap=matplotlib.pyplot.cm.cool)
```
