## clustering Week5  

```python
import pandas as pd
import numpy as np
import sklearn.cluster
import matplotlib.pyplot
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# import data from excel
data = pd.read_excel('GA_Data_current.xls')
data
```

```python
# drop string columns and sort 
#data.drop(['Supplier','Ref_No','Date'], axis=1, inplace=True)
data.sort_values('Robot_Strength', inplace=True)
```

```python
data['cost_per_strength']= data['Supplier_Cost']/data['Robot_Strength']
```

```python
plt.scatter(data.Robot_Strength, data.Supplier_Cost)
```

```python
k_means = sklearn.cluster.KMeans (init ='k-means++',n_clusters=3 )
target_info = data[['Robot_Strength','Supplier_Cost']]
```

```python
k_means_cs = k_means.fit_predict(target_info)
data['k_means_cs']= k_means_cs
```

```python
Colouring = np.array(['red','green','blue'])
plt.scatter(data.Robot_Strength, data.Supplier_Cost,
                          color=Colouring[k_means_cs])

plt.scatter(k_means.cluster_centers_[:,0],
                          k_means.cluster_centers_[:,1],
                          marker="+", s=30)
```

```python
dbscan = sklearn.cluster.DBSCAN(eps=30, min_samples=10)
```

```python
target_info
```

```python
#dbscan_cs= dbscan.fit_predict(target_info)

#data['dbscan_cs']=dbscan_cs
dbscan_cs= dbscan.fit(target_info)
```

```python
core_samples_mask = np.zeros_like(dbscan_cs.labels_, dtype=bool)

core_samples_mask[dbscan_cs.core_sample_indices_] = True
labels = dbscan_cs.labels_
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_clusters_
```

```python
Colouring = np.array(['red','green','blue'])
plt.scatter(data.Robot_Strength, data.Supplier_Cost,
                          color=Colouring[dbscan_cs])

#plt.scatter(dbscan.cluster_centers_[:,0],
#                          dbscan.cluster_centers_[:,1],
 #                         marker="+", s=30)
```

## Ensemble - Bagging Regression

```python
import sklearn.ensemble
```

```python
bag = sklearn.ensemble.BaggingRegressor(max_features=3, n_estimators=25, max_samples=30, n_jobs=1)
```

```python
robot_predictors = data.drop(['Supplier_Cost','cost_per_strength','Final_Cost'], axis=1)
```

```python
bag.fit(robot_predictors,data.Supplier_Cost)
```

```python
matplotlib.pyplot.scatter(data.Robot_Strength, data.Supplier_Cost)
matplotlib.pyplot.plot(data.Robot_Strength, bag.predict(robot_predictors))
```

## Random Forest Regression 

```python

rft= sklearn.ensemble.RandomForestRegressor(n_estimators = 30)
```

```python
rft.fit(robot_predictors, data.Supplier_Cost)
```

```python
matplotlib.pyplot.scatter(data.Robot_Strength, data.Supplier_Cost)
matplotlib.pyplot.plot(data.Robot_Strength, rft.predict(robot_predictors))
```
