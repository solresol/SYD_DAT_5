```python
import pandas as pd
import numpy as np
import sklearn.cluster
import matplotlib.pyplot
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# import data from excel
data = pd.read_excel('GA_Data_V1.xls')
data
```

```python
# drop string columns and sort 
#data.drop(['Supplier','Ref_No','Date'], axis=1, inplace=True)
data.sort_values('Robot_Strength', inplace=True)
```

```python
data
```

```python
data['cost_per_strength']= data['Supplier_Cost']/data['Robot_Strength']
```

```python
plt.scatter(data.Robot_Strength, data.Supplier_Cost)
```

## K-Means Clustering for Robot_Strength and Cost 

```python
k_means = sklearn.cluster.KMeans (init ='k-means++',n_clusters=3 )
StrengthNCost = data[['Robot_Strength','Supplier_Cost']]
```

```python
k_means_StrengthVSCost = k_means.fit_predict(StrengthNCost)
data['k_means_StrengthVSCost']=k_means_StrengthVSCost
```

```python
Colouring = np.array(['red','green','blue'])
plt.scatter(data.Robot_Strength, data.Supplier_Cost,
                          color=Colouring[k_means_StrengthVSCost])

plt.scatter(k_means.cluster_centers_[:,0],
                          k_means.cluster_centers_[:,1],
                          marker="+", s=30)

sklearn.metrics.silhouette_score(StrengthNCost, k_means.labels_, metric='euclidean')
```

## k-means clustering for supplier cost and cost per strength

```python
supplierCost_and_CPS = data[['Supplier_Cost','cost_per_strength']]
k_means_supplierCostVsCPS = k_means.fit_predict(supplierCost_and_CPS)
plt.scatter(data.Supplier_Cost, data.cost_per_strength, color = Colouring[k_means_supplierCostVsCPS])
plt.scatter(k_means.cluster_centers_[:,0], k_means.cluster_centers_[:,1], marker ="+", s=30 )

sklearn.metrics.silhouette_score(supplierCost_and_CPS, k_means.labels_, metric='euclidean')
```

## Ensemble - Bagging Regression

```python
import sklearn.ensemble
```

```python
bag = sklearn.ensemble.BaggingRegressor(max_features=3, n_estimators=50, max_samples=100, n_jobs=1)
```

```python
robot_predictors = data.drop(['Supplier_Cost','cost_per_strength','Final_Cost'], axis=1)
```

```python
bag.fit(robot_predictors,data.Supplier_Cost)
```

```python
matplotlib.pyplot.scatter(bag.predict(robot_predictors), data.Supplier_Cost)
#matplotlib.pyplot.plot(data.Robot_Strength, bag.predict(robot_predictors))
```
