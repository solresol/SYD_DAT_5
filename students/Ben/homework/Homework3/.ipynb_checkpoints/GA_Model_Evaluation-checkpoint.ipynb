## Model Evaluation 

```python
import pandas as pd
import numpy as np
import sklearn.cross_validation
import sklearn.linear_model
import sklearn.metrics
import sklearn.tree
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

## setup and run lasso model

```python
data = pd.read_excel('GA_Data_current_v1.xls', convert_float=False)
data_sorted = data.sort_values('Robot_Strength', axis = 0)
Robot_Strength = data.Robot_Strength.reshape(-1,1)
```

```python
data
```

```python
# try with lasso model with 
lasso = sklearn.linear_model.Lasso(alpha = 0.1, max_iter=10000)
lasso.fit(Robot_Strength, data.Supplier_Cost)
prediction_lasso = lasso.predict(Robot_Strength)
plt.scatter(data.Robot_Strength, data.Supplier_Cost)
plt.plot(data.Robot_Strength, prediction_lasso)
```

```python
features = data_sorted.drop(['Ref_No','Date','Supplier','Supplier_Cost'], axis = 1)
features
```

```python
lasso.fit(features, data_sorted.Supplier_Cost)
prediction_lasso_overfit=lasso.predict(features)
plt.scatter(data_sorted.Robot_Strength, data_sorted.Supplier_Cost)
plt.plot(data_sorted.Robot_Strength, prediction_lasso_overfit)
```

## split and train data with simple lasso model


```python
x = data_sorted.Robot_Strength
y = data_sorted.Supplier_Cost
```

```python
split_data = sklearn.cross_validation.train_test_split(x,y, test_size = 0.3)
split_data
```

```python
x_train= split_data[0]
x_test= split_data[1]
y_train = split_data[2]
y_test = split_data[3]
```

```python
lasso_test = sklearn.linear_model.Lasso(alpha = 0.1, max_iter=1000)
lasso_test.fit(x_train.reshape(-1,1), y_train)
#prediction_lasso = lasso.predict(Robot_Strength)
#plt.scatter(data.Robot_Strength, data.Supplier_Cost)
#plt.plot(data.Robot_Strength, prediction_lasso)

#lasso.fit(x_train, y_train)
```

```python
y_prediction = lasso_test.predict(x_test.reshape(-1,1))
```

```python
variance = sklearn.metrics.explained_variance_score(y_test,y_prediction)
mean_absolute_error = sklearn.metrics.mean_absolute_error(y_test,y_prediction)
median_absolute_error = sklearn.metrics.median_absolute_error(y_test,y_prediction)
r2 = sklearn.metrics.r2_score(y_test,y_prediction)
```

```python
lasso_test.score(x_test.reshape(-1,1), y_test)
```

```python
print (variance, mean_absolute_error,median_absolute_error,r2)
```

```python
variance = sklearn.metrics.explained_variance_score(y_test_overfit,y_prediction_overfit)
mean_absolute_error = sklearn.metrics.mean_absolute_error(y_test_overfit,y_prediction_overfit)
median_absolute_error = sklearn.metrics.median_absolute_error(y_test_overfit,y_prediction_overfit)
r2 = sklearn.metrics.r2_score(y_test_overfit,y_prediction_overfit)

print (variance, mean_absolute_error,median_absolute_error,r2)
```
